```python
def bangla_char_in_word(word):
    num_bangla_char = 0
    for char in word:
        if (ord(char) >= 0x0980  and ord(char)<=0x09FF) == True:
            num_bangla_char += 1
    return num_bangla_char
```

```python
def is_bangla_sentence (text):
    #first split it with the space
    tot_char_count = 0
    bangla_char_count = 0
    words = text.split()
    for word in words:
        tot_char_count += len(word)
        bangla_char_count += bangla_char_in_word(word)
    #print(tot_char_count,bangla_char_count)
    if(bangla_char_count >= (0.4 * tot_char_count)):
        return True
    return False
```

```python
def remove_punctuation(s):
    
 s = re.sub(r'[^\w\s]','',s)
 return s
```

```python
def is_english_text(text):
    #first remove the punctuation
    text = remove_punctuation(text)
    #Then split it on whitespace
    words = text.split()
    #And check each word one by one
    total_word_count = len(words)
    english_word_count = 0
    for word in words:
        
        if(word in english_words):
            english_word_count += 1
    
    if(english_word_count >= 0.6*total_word_count):
        return True
    return False
```

```python
def write_as_words(words_in_post, file_object):
    if(len(words_in_post) == 0):
        return
    for i in range(len(words_in_post)-1):
        file_object.write(words_in_post[i] +" ")
    file_object.write(words_in_post[-1] + "\n")
     
```

```python
#this method reads the csv file rows one by one and separates them into three different file for each of Bangla, Benglish and English status

def separate_status_by_language():
    all_status = pd.read_csv("./uber_group_posts.csv")
    all_status.columns
    all_status = all_status[["_id","status_message"]]
    stop_words = set(stopwords.words('english'))
    
    domain_words = ['uber','driver','car','trip','ride','fare','time','call','come','go','get','drivers']
 
    
 

    output = open("status_by_language.txt","w+")
    
    bangla_data = open(sample_data_path + "bangla_dataset.txt","w+")
    english_data = open( sample_data_path + "english_dataset.txt","w+")
    banglish_data = open(sample_data_path + "banglish_dataset.txt","w+")
    print(len(all_status))
    ps =   WordNetLemmatizer()
  
   
    count = 0
    #random_indices = np.random.random_integers(0, len(all_status),200)
    #print(sorted(random_indices))
    for i in range(len(all_status)):
        #print(i)
        cur_status = all_status.iloc[i]["status_message"]
        
         #first check whether the status is actually string
        if(isinstance(cur_status,str) == False):
            continue
        count += 1
        #output.write(cur_status)
        table = str.maketrans({key: ' ' for key in string.punctuation})
        cur_status = cur_status.translate(table) 
        #output.write(cur_status)
        
        #print(count)
        #output.write(str(i)+":\n")
        #output.write(cur_status)
        
 
        word_tokens = word_tokenize(cur_status)
        words_in_post = cur_status.split()
 
        
        if(is_bangla_sentence(cur_status)):
            #print("Bangla")
            #output.write("\n output: Bangla")
            
            write_as_words(words_in_post, bangla_data)

        elif(is_english_text(cur_status)):
            #print("English")
            #output.write("\n output: English")
            #print(word_tokens)
            print(str(count)+"\n")
            output.write(' '.join(word_tokens))
            #print(word_tokens)
            filtered_sentence = [ps.lemmatize(w.lower(),'v') for w in word_tokens if w.lower() not in stop_words and ps.lemmatize(w.lower(),'v') not in domain_words]
            print(filtered_sentence)
            #output.write(filtered_sentence))
          
            #print(filtered_sentence)
            write_as_words(filtered_sentence, english_data)
        else:
            #print("Banglish")
            #output.write("\n output: Banglish")
            write_as_words(words_in_post, banglish_data)
        #print("\n\n")
        output.write("\n\n")
    output.close()
    print("count: " +str(count))
    return all_status
    
    
```
